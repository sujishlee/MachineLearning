# Hierarchical Clustering

##### Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Loading the dataset

```python
file_path='C17_HierarchicalClustering_Mall_Customers.csv'
dataset = pd.read_csv(file_path)
print(dataset)
```

####  Creating the matrix of features (X) 

```python
X=dataset.iloc[:,[3,4]].values
print(X)
```

#### Using the dentrogram tofind the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dentogram=sch.dendrogram(sch.linkage(X,method='ward'))


plt.title('Dentrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

#### Training the K-Means model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```
